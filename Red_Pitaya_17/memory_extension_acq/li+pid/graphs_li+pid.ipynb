```python
%matplotlib inline

from math import *
from pylab import *

import numpy as np
import matplotlib.pyplot as plt

# import data
import original_dec
import extended_dec
```

```python
BITS_PER_SAMPLE = 32        # bits
MAX_SAMPLE_RATE = 125e6     # Hz
```

### 1. Original (No Memory Extension)

##### 1.1 Calculations

```python
maxMemBytes = 16**4
maxNumMemSamples = maxMemBytes * 8 / BITS_PER_SAMPLE

print('Per channel:')
print('Max memory allocated = {} bytes'.format(maxMemBytes))
print('Max number of samples = {}'.format(int(maxNumMemSamples)))
```

```python
signalPeriod = maxNumMemSamples / MAX_SAMPLE_RATE
signalFrequency = 1 / signalPeriod

print('Singal period = {} μs'.format(signalPeriod * 1e6))
print('Signal frequncy = {} Hz'.format(signalFrequency))
```

#### 1.2 Data & Graphs

```python
nSamples = len(original_dec.ch1)
tSamples = original_dec.t1[-1] - original_dec.t1[0]
totDecFactor = maxNumMemSamples / nSamples

print('Sampling time = {} μs'.format(tSamples))
print('Number of samples saved = {}'.format(nSamples))
print('Total decimation factor = {}'.format(totDecFactor))
```

```python
plt.figure(figsize=(7.5, 5))

plt.plot(original_dec.t1, original_dec.ch1, 'r-', lw=1, label="signal")

plt.title('Original (No Memory Extension)', fontsize=15)
plt.xlabel(r'Time ($\mu$s)', fontsize=15)
plt.ylabel('Voltage (V)', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

legendTitle = 'N = {}'.format(nSamples)
plt.legend(title=legendTitle, loc='best', title_fontsize='x-large', prop={'size': 15})

plt.show()
```
